```python
import numpy as np
import math
```

```python
P = np.array([[0,0], [0.2, 0.22], [0.1, -0.1], [1,1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63], [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8,-0.23], [0.8, 0.6]])
k = 3 #given values if we vary this we can get that much nearest neighbours for eacch point
```

```python
def k_nearest_neighbors(points, k):
   pairwise_distances = np.sqrt(np.sum((points[:, np.newaxis] - points) ** 2, axis=2)) #euclidean distance
   nearest_k = np.argsort(pairwise_distances, axis=1)[:, 1:k+1] #sorting based on euclidean distance
   return nearest_k
```

```python
k_nearest_neighbors(P,k) #finding nearest neighbours for each
```

```python

import matplotlib.pyplot as plt
nearest_indices = k_nearest_neighbors(P, k)
plt.scatter(P[:, 0], P[:, 1], c='b', label='Points')
for i in range(len(P)):
  for j in nearest_indices[i]:
    plt.arrow(P[i, 0], P[i, 1], P[j, 0] - P[i, 0], P[j, 1] - P[i, 1],linewidth=2,alpha=0.7, fc='g', ec='g',head_width=0.05, head_length=0.05,length_includes_head=True)
#using matplotlibs arrow to plot the in the scatter


plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
