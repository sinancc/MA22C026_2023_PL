```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#Plotting Sine curve

```python
x=np.linspace(0,10) #taking points for graphing
```

```python
y=np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero') #removing unnecessary spines
ax.spines['left'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
plt.plot(x,y,'b-',label="sine function") #plotting sine on points
plt.title("graph of sine function") #adding title,labels and kegend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

#3D plotting

```python
from mpl_toolkits.mplot3d import Axes3D #importing necessary modules
```

```python
def func(x, y): #defining function
    return np.cos(np.sqrt(x**2 + y**2))
```

```python
x=np.linspace(-5,5)
y=np.linspace(-5,5)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
Z = func(X, Y)
```

```python
fig = plt.figure(figsize=(8, 8))
p = fig.add_subplot(111, projection='3d')
p.plot_surface(X, Y, Z, cmap='viridis')# graphing function
p.set_xlabel('x') #labelling axes
p.set_ylabel('y')
p.set_zlabel('function values')
p.set_title('Plot of the function')#setting Title
plt.show()
```
