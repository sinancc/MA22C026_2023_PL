```python
a=time.time() #start time
x = np.linspace(-5, 5, 1000001) #uniformly spaced 1000001 nodes in [-5,5]
y = np.where(x < 0, x * np.sin(x**2), np.cos(x)) #f(x)
b=time.time() #end time
```

```python

plt.plot(x, y,'r-',label="f(x)") #matplotlib for ploting graph
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x)')
plt.legend()
plt.show()
print("time taken for this process is:",b-a) #time
```
