```python
#necessary modules
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import datetime
```

```python

# Define the function f(x, t)
def f(x, t):
    return 0.5 * (x + t)**2 + 2 * np.sin(10 * (x - t))
```

```python


# Create a figure and axis for plotting
fig, ax = plt.subplots()
x_values = np.linspace(0, 10, 1000)
line, = ax.plot(x_values, f(x_values, 4))  # Initialize the plot with t = 4

# Define the animation function
def animate(t):
    ax.clear()
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 150)  # Adjust the y-axis limits if needed
    line, = ax.plot(x_values, f(x_values, t))
    ax.set_title(f'f(x, t) for t = {t:.1f}')

# Create the animation
%timeit animation = FuncAnimation(fig, animate, frames=np.linspace(0, 10, 100), interval=50)

# Display the animation
plt.show()

```
