#Importing Data Required

```python
# Download and preprocess COVID-19 data from JHU-CSSE
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
url2= 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
```

#Essential data Munging

```python

data = pd.read_csv(url)
data.head()
```

```python
data2=pd.read_csv(url2)
data2.head()
```

```python
# Drop un necessary columns  in the DataFrame
data2= data2.rename(columns={'iso3': 'iso_a3'})
columns_to_drop = ['UID', 'iso2', 'code3', 'FIPS', 'Admin2' ,'Combined_Key','Lat','Long_']
data2 = data2.drop(columns=[col for col in columns_to_drop if col in data2.columns])
```

```python
data2.head()
```

```python
data2=data2.groupby('iso_a3').sum()
data2.head()
```

```python
data2= data2.rename(columns={'iso3': 'iso_a3'})
```

```python
data2.head()
```

```python
len(data2)
```

```python
# make a geometry object from Lat, Long
geo = gpd.points_from_xy(data['Long'], data['Lat'])
# give the geometry to geopandas together with data
gdata = gpd.GeoDataFrame(data, geometry=geo)
gdata.head()
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world= world.rename(columns={'name': 'Country/Region'})
world.head()
```

```python

latest_data = pd.merge(world,country_data_covid, on='Country/Region',how='outer')
```

```python
# plot coloring counties by number of confirmed cases
fig, ax = plt.subplots(figsize=(12, 8))
t='3/9/23'
world.boundary.plot(ax=ax, linewidth=1, color='k')  # Add country borders
latest_data_2.plot(ax=ax, column=f'{t}_x', legend=True, cmap='YlOrRd',legend_kwds={'label': '# confimed cases', 'orientation':'horizontal'})
latest_data_2.plot(ax=ax,cmap='YlOrRd', column=f'{t}_y')

# label the counties
# for x,y,Country in zip(latest_data['Lat'], latest_data['Long'],latest_data['Country/Region']): ax.text(x,y, Country, color='black',size=5)
ax.set_title(f'Confirmed COVID-19 cases countrywise as of {t}')
ax.set_xlabel('Latitude'); ax.set_ylabel('Longitude');
plt.savefig('covid19_world_map.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation
from celluloid import Camera
```

```python
from datetime import datetime, timedelta

start_date = datetime(2020, 1, 22)
end_date = datetime(2023, 3, 9)

date_list = []

current_date = start_date
while current_date <= end_date:
    formatted_date = current_date.strftime('%-m/%-d/%y')  # Remove zero padding
    date_list.append(formatted_date)
    current_date += timedelta(days=1)
print(date_list)


```

```python
len(date_list)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
camera = Camera(fig)
for i in range(0,1143,15):
    t = date_list[i]
    world.boundary.plot(ax=ax, linewidth=1, color='k')  # Add country borders
    latest_data_2.plot(ax=ax, column=f'{t}_x', legend=False, cmap='YlOrRd')
    latest_data_2.plot(ax=ax, cmap='YlOrRd', legend=False, column=f'{t}_y')
    cax = fig.add_axes([0.2, 0.05, 0.6, 0.02])
    sm = plt.cm.ScalarMappable(cmap='YlOrRd')

    plt.text(3, 1, f'Date={t}', fontsize=12, color='red',position=(1,0))
    sm.set_array([])
    cbar = plt.colorbar(sm, cax=cax, orientation='horizontal')
    camera.snap()

# Create the animation
animation = camera.animate()

# Save the animation to a file (e.g., MP4)
animation.save('covid_world_animation.mp4', writer='ffmpeg',fps=3)
plt.show()
```
