```python
import numpy as np
```

```python
def create_matrix(n, k):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(max(0, i - k), min(n, i + k + 1)):
            A[i, j] = 1
    return A
```

```python
 create_matrix(15,4)
```

```python
def averaging_operator(n,k):
  A=create_matrix(n,k)
  A /= np.sum(A,axis=1, keepdims=True)
  return A
```

```python
averaging_operator(15,4)
```

```python
import numpy as np
import math
n = 1000
k=200
x = np.linspace(0,10,n)
def function_value(x):
    return x**2 + 2 * np.sin(10 * x)

y= function_value(x)
A=averaging_operator(n,k)
np.dot(y,A)
```
