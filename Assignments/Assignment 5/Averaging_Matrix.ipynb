```python
import numpy as np
```

```python
def create_matrix(n, k):
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(max(0, i - k), min(n, i + k + 1)):
            A[i, j] = 1
    return A
```

```python
 create_matrix(15,4)
```

```python
def averaging_operator(n,k):
  A=create_matrix(n,k)
  A /= np.sum(A,axis=1, keepdims=True)
  return A
```

```python
averaging_operator(15,4)
```

```python
import numpy as np
import math
n = 20
k=2
def vec(n):
  x=np.zeros(n)
  for i in range(n):
    x[i]=(-1)**i
  return x

y= vec(n)
A=averaging_operator(n,k)
np.dot(A,y)
```
