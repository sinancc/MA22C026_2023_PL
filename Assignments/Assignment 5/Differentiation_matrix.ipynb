```python
def matrix_D(n):
  A= np.eye(n-1, n, k=1) + np.eye(n-1, n,k=0) *-1
  return A
```

```python
n=10
A=matrix_D(n)
A
```

```python
B = np.random.rand(n)
B
```

```python
np.dot(A,B)
```

#f(x)=sin(x) deriavtive approximation

```python
import math
x_j=np.linspace(0,2*math.pi,1000)
y=np.sin(x_j)
y_1=np.dot(matrix_D(1000),y)
y_2=y_1/(2*3.14/1000)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_j,y,label="sin")
plt.plot(x_j[1:],y_2,label="derivative",linewidth=3)
z=np.cos(x_j)
plt.plot(x_j[1:],z[1:],label="cos",linewidth=1)
plt.legend()
```

Derivative of sin(x) is cos(x).which we gets by diffrentiation matrix

#Derivative of f (x)= x^2 +2 sin(10x)

```python
import math
x_k=np.linspace(0,2*math.pi,1000)
y_new=2*np.sin(10*x_k)+x_k**2
y_a=np.dot(matrix_D(1000),y_new)
y_b=y_a/(2*3.14/1000)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_k,y_new,label="function")
plt.plot(x_k[1:],y_b,label="approximation of derivative",linewidth=5)
z=2*x_k+2*10*np.cos(10*x_k)
plt.plot(x_k,z,label="derivative")
plt.legend()
```
