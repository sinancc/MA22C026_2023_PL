```python
import pandas as pd
%matplotlib inline
import yfinance as yf
import matplotlib.pyplot as plt
```

#Data importing

```python
from operator import index
symbols=['AAPL','GOOG','TSLA']
s = yf.download(symbols, start='2023-01-01')
s.head()
```

```python
if isinstance(s.index, pd.MultiIndex):
    print("The DataFrame has a MultiIndex.")
else:
    print("The DataFrame has a regular index.")
```

#Closing Prices of AAPL

```python
aapl_closing_prices = s['Adj Close']['AAPL']

# Plot the closing prices
plt.figure(figsize=(12, 6))
plt.plot(aapl_closing_prices)
plt.title("AAPL Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid()
plt.show()
```

```python
# Select the 'Adj Close' column for all symbols and use .iloc[-1] to get the last (most recent) row.
closing_prices_yesterday = s['Adj Close'].iloc[-1]

print("Closing Prices Yesterday:")
print(closing_prices_yesterday)
```

#TSLA Data Alone

```python
# Select the 'Adj Close' column for TSLA
tsla_data = s['Adj Close']['TSLA']

print("TSLA Data:")
print(tsla_data)
```
