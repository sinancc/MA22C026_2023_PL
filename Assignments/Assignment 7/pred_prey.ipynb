#plotting of the graph for 0<=t<=70 and r(0)=5 &x(0)=2

```python
def pred_prey(t, y, alpha,beta,delta,gmmma):
  r,x = y
  return np.array([- beta * r * x + alpha * r ,- gamma*x+delta* r *x ])
```

```python
alpha=1.1 #Given values of args
beta = 0.4
delta = 0.1
gamma = 0.1
```

```python
sol = solve_ivp(pred_prey, [0, 70], [5,2], rtol=1e-6, args=(alpha,beta,delta,gamma)) # solve ivp for solving initial value problems
```

```python
sol
```

```python
fig = plt.figure(); ax = fig.gca() #plotting of graph
curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['Rabbit', 'Fox'])
ax.set_xlabel('Time (t)')
ax.set_ylabel('Population')
ax.set_title('Predator Prey model for Rabbits and Foxes')
ax.grid()

plt.show()
```
