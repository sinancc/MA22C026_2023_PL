#plotting of the graph for 0<=t<=70 and r(0)=5 &x(0)=2

```python
def pred_prey(t, y, alpha,beta,delta,gmmma):
  r,x = y
  return np.array([- beta * r * x + alpha * r ,- gamma*x+delta* r *x ])
```

```python
alpha=1.1 #Given values of args
beta = 0.4
delta = 0.1
gamma = 0.1
```

```python
sol = solve_ivp(pred_prey, [0, 70], [5,2], rtol=1e-6, args=(alpha,beta,delta,gamma)) # solve ivp for solving initial value problems
```

```python
sol
```

```python
fig = plt.figure(); ax = fig.gca() #plotting of graph
curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['Rabbit', 'Fox'])
ax.set_xlabel('Time (t)')
ax.set_ylabel('Population')
ax.set_title('Predator Prey model for Rabbits and Foxes')
ax.grid()

plt.show()
```

#Prepare a fgure (phase portrait) with phase plots of, say, 10 solutions, one each for randomly chosen initial values r(0) and x(0) between 1 and 9.

```python
rand_init_values=np.random.uniform(1, 9, (10, 2)) #random values between 1 and 9.2 each for 10 solutions
```

```python
plt.figure(figsize=(7,7))
for i in range(10):
    r0, x0 = rand_init_values[i] #taking one elemnt from above array
    sol = solve_ivp(pred_prey, (0, 70), [r0, x0], rtol=1e-12, args=(alpha,beta,delta,gamma)) #using predefined pred_prey
    plt.plot(sol.y[1], sol.y[0], label=f'solution {i+1}') #plotting
```
